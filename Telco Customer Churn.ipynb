## Data Preprocessing

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors ="coerce")
```

```python
df.TotalCharges.dtypes
```

```python
df.isnull().sum()
```

### Handling Missing Data

```python
df.TotalCharges = df.TotalCharges.fillna(0)
```

```python
df.isnull().sum().sum()
```

### Dealing with Columns and Values in the Dataset

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
```

### Dealing with Values in the Columns

```python
string_columns = list(df.dtypes[df.dtypes == "object"].index)
```

```python
string_columns
```

```python
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(" ","_")
```

```python
df.head()
```

### Addressing the target and features

```python
df.churn.head()
```

```python
df.churn = (df.churn == "yes").astype(int)
```

```python
df.churn.head()
```

```python
df.head()
```

### Splitting into Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
df_train_full, df_test = train_test_split(df, test_size = 0.2, random_state = 42)
```

```python
df_train, df_val = train_test_split(df_train_full, test_size = 0.25, random_state = 42)
```

```python
y_train = df_train.churn.values
```

```python
y_val = df_val.churn.values
```

```python
del df_train["churn"]
```

```python
del df_val["churn"]
```

### Determining Categorical and Numerical Features

```python
categorical = ['gender', 'seniorcitizen', 'partner', 'dependents', 'phoneservice',
               'multiplelines', 'internetservice', 'onlinesecurity', 'onlinebackup',
               'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies',
               'contract', 'paperlessbilling', 'paymentmethod']
```

```python
numerical = ['tenure', 'monthlycharges', 'totalcharges']
```

### One-Hot Encoding

```python
train_dict = df_train[categorical + numerical].to_dict(orient = "records")
```

```python
train_dict[:1]
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse = False)
```

```python
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
```

```python
X_train[0]
```

```python
dv.get_feature_names_out()
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver = "liblinear", random_state = 42)
```

```python
model.fit(X_train, y_train)
```

## Model Evaluation

```python
val_dict = df_val[categorical+numerical].to_dict(orient="records")
```

```python
X_val = dv.transform(val_dict)
```

```python
y_pred = model.predict_proba(X_val)
```

```python
y_pred[:5]
```

```python
print("The performance of the model on the validation dataset: ", model.score(X_val, y_val))
print("The performance of the model on the training dataset: ", model.score(X_train, y_train))
```

### Coefficients of the Model

```python
model.score(X_val, y_val)
```

```python
model.score(X_train, y_train)
```

```python
model.intercept_[0]
```

```python
dict(zip(dv.get_feature_names_out(), model.coef_[0].round(3)))
```
